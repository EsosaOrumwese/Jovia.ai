### Branching with **`if`**, **`else`** and **`elif`**
**`IF` statement**

```python
a_number = 34

if a_number % 2 == 0:
      print("We're inside an if block")
      print("The given number {} is even.".format(a_number))
```

```python
a_number = 33

if a_number % 2 == 0:
      print("The given number {} is even.".format(a_number))
```

```python
a_number = 33

if a_number % 2 == 0:
      print('The given number {} is even.'.format(a_number))
else:
      print('The given number {} is odd.'.format(a_number))
```

Iteration: **`While`** loop

```python
result = 1
i = 1

while i <= 100:
      result = result * i
      i = i+1
      
print('The factorial of 100 is: {}'.format(result))
```

Let's see how long it takes to run this iteration using the **%%time** command at the top of the cell

```python
%%time

result = 1
i = 1

while i <= 1000:
      result *= i
      i += 1
      
print('The factorial of 100 is: {}'.format(result))
```

```python
line = '*'
max_length = 10

while len(line) < max_length:
      print(line)
      line += '*' # adds an extra asterisk to at the end

while len(line) > 0:
      print(line)
      line = line[:-1] # omits the last asterisk in the 'line' variable
```

```python
marker = '*'
max_length = 10
space = " " * max_length
i = 1

while i < max_length:
      line = space[:-i] + marker * i
      print(line)
      i += 1

while i > 0:
      i -= 1
      space = " " * (max_length-i)
      line = space + marker * (i)
      print(line)
      
```

```python
marker = '*'
end = marker
max_length = 10
space = " " * max_length
i = 1

while i < max_length:
      line = space[:-i] + marker * i + end * (i-1)
      print(line)
      i += 1
      
while i > 0:
      i -= 1
      space = " " * (max_length-i)
      line = space + marker * (i) + end * (i-1)
      print(line)
```

**`break`** and **`continue`** statements

```python
i = 1
result = 1

while i <= 100:
      result *= i
      if i == 42:
            print('Magic number 42 reached! Stopping execution..')
            break
      i += 1

print('i:', i)
print('result:', result)
```

```python
i = 1
result = 1

while i < 20:
      i += 1
      if i % 2 == 0:
            print('Skipping {}'.format(i))
            continue
      print('Multiplying with {}'.format(i))
      result *= i

print('i:', i)
print('result:', result)
```

Ranges are used for iterating over lists when you need to track the index of elements while iterating.

```python
a_list = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

for i in range(len(a_list)):
      print('The value at position {} is {}.'.format(i, a_list[i]))
```

```python
for i, val in enumerate(a_list):
      print('The value at position {} is {}.'.format(i, val))
```

**`break`**, **`continue`** and **`pass`** statements

```python
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
```

```python
for day in weekdays:
      print('Today is {}'.format(day))
      if (day == 'Wednesday'):
            print("I don't work beyond Wednesday!")
            break
```

```python
for day in weekdays:
      if (day in ['Wednesday', 'Friday']):
            print("I don't work on {}".format(day))
            continue
      print('Today is {}'.format(day))
```
